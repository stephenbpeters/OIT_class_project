```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
print("libraries imported!")
```

```python
df = sns.load_dataset("penguins")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.histplot(df["flipper_length_mm"], kde=True)
plt.title("Distribution of Flipper Lengths")
plt.show()
```

```python
sns.scatterplot(data=df, x="bill_length_mm", y="bill_depth_mm", hue="species")
plt.title("Bill Length vs Bill Depth by Species")
plt.show()
```

```python
df.groupby("species")[["bill_length_mm", "flipper_length_mm"]].mean()
```

```python
import ipywidgets as widgets
from IPython.display import display

species_selector = widgets.Dropdown(options=df["species"].dropna().unique(), description="Species:")

def show_species_summary(species):
    display(df[df["species"] == species].describe())

widgets.interact(show_species_summary, species=species_selector)
```
