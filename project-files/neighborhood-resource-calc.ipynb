```python
# first let's sum all the resources except libraries
resource_cols = [
    "temporary_shelter", "day_center", "coord_reentry", "group_therapy",
    "showers", "meals", "food_box", "laundry", "health_services", "misc_services"
]

yn_map = {"yes": 1, "y": 1, "true": 1, "t": 1, "no": 0, "n": 0, "false": 0, "f": 0}
df[resource_cols] = (df[resource_cols]
    .apply(lambda s: s.str.lower().map(yn_map) if s.dtype == "object" else s)
    .apply(pd.to_numeric, errors="coerce")
    .fillna(0))

df["resource_calc-nolib"] = df[resource_cols].sum(axis=1).astype(int)

df
```

```python
# now we create resource_calc_total which adds one to our resource score if there is a library and it is currently open

import numpy as np

df["resource_calc_total"] = df["resource_calc-nolib"] + (
    ((df["library"] > 0) & (df["open"] > 0)).astype(int)
)

df
```

```python
# import pandas as pd

# load our file:
base_dir = Path("C:/Users/Steph/local/OIT-class/project-files/datasets/processed")
df_start = pd.read_csv(base_dir / "211_Info_Data_latlong_hoods_all-csv.csv")

# Work on a copy and tidy column names
df_start = df_start.copy()
df_start.columns = df_start.columns.str.strip()

# Ensure neighborhood is present and usable
df_start = df_start.dropna(subset=["neighborhood"])
df_start["neighborhood"] = df_start["neighborhood"].astype("string").str.strip()

# Columns to aggregate
service_cols = [
    "temporary_shelter", "day_center", "coord_reentry", "group_therapy",
    "showers", "meals", "food_box", "laundry", "health_services", "misc_services"
]
agg_cols = ["library_open"] + service_cols

# If any expected column is missing, create it as zeros
for c in agg_cols:
    if c not in df_start.columns:
        df_start[c] = 0

# Coerce to numeric (0/1 etc.); treat non-numeric/NaN as 0
df_start[agg_cols] = (df_start[agg_cols]
    .apply(pd.to_numeric, errors="coerce")
    .fillna(0)
    .astype(int)
)

# Sums per neighborhood for those columns
sums_per_hood = df_start.groupby("neighborhood")[agg_cols].sum()

# Count of rows per neighborhood
counts_per_hood = df_start.groupby("neighborhood").size().rename("number_of_locations")

# Combine
df_hoods = sums_per_hood.join(counts_per_hood)

# final_resource_calc = sum of (library_open + the 10 service columns)
df_hoods["final_resource_calc"] = df_hoods[agg_cols].sum(axis=1)

# Make neighborhood a column and (optionally) sort
df_hoods = df_hoods.reset_index().sort_values(
    ["final_resource_calc", "neighborhood"], ascending=[False, True]
).reset_index(drop=True)

print(df_hoods.head())
```

```python
# ok, let's save to .csv:
base_dir_save = Path("C:/Users/Steph/local/oit-class/project-files/datasets/processed")

# Ensure the directory exists (create it if it doesn't)
base_dir_save.mkdir(parents=True, exist_ok=True)

# Define the full output file path
output_file = base_dir_save / "resources_by_hood_calc-csv.csv"

# Save the DataFrame to CSV
df_hoods.to_csv(output_file, index=False)

print(f"File saved to: {output_file}")
print("This file has the count of each kind of resource per neighborhood...")
print("And it has the sum of all the resources per neighborhood...")
print("And the number of individual resource locations per neighborhood.")
print("It would be good to have someone else proofread this, but I think it is good.")
```
